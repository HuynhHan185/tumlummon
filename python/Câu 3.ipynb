```python
class Media:
    def __init__(self, id, title, year, category, available):
        self.id = id
        self.title = title
        self.year = year
        self.category = category
        self.available = available

    def __str__(self):
        return f"ID: {self.id}, Title: {self.title}, Year: {self.year}, Category: {self.category}, Available: {self.available}"
```

```python
class Book(Media):
    def __init__(self, id, title, year, category, available, author, publisher, pages, ISBN):
        super().__init__(id, title, year, category, available)
        self.author = author
        self.publisher = publisher
        self.pages = pages
        self.ISBN = ISBN

    def __str__(self):
        return super.__str__() + f"Author: {self.author}, Publisher: {self.publisher}, Pages: {self.pages}, ISNB: {self.ISNB}"
```

```python
class DVD(Media):
    def __init__(self, id, title, year, category, available, director, runtime, language, subtitles):
        super().__init__(id, title, year, category, available)
        self.director = director
        self.runtime = runtime
        self.language = language
        self.subtitles = subtitles

    def __str__(self):
        return super.__str__() + f"Director: {self.director}, Runtime: {self.runtime}, Language: {self.language}, Subtitles: {self.subtitles}"
```

```python
class Journal(Media):
    def __init__(self, id, title, year, category, available, volume, issue, publisher, field):
        super().__init__(id, title, year, category, available)
        self.volume = volume
        self.issue = issue
        self.publisher = publisher
        self.field = field

    def __str__(self):
        return super.__str__() + f"Volume: {self.volume}, Issue: {self.issue}, Publisher: {self.publisher}, Field: {self.field}"
```
